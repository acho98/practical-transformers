### 개요
모바일버트(MobileBERT)는 BERT(Bidirectional Encoder Representations from Transformers)의 경량화된 버전으로, 주로 모바일 기기와 같이 계산 자원이 제한된 환경에서 사용될 수 있도록 최적화된 모델이다. BERT와 동일한 자연어 처리 작업(NLP)에서 성능을 유지하면서도 훨씬 가벼운 모델을 목표로 설계되었다.

### 특징
- 경량화된 구조: MobileBERT는 BERT와 유사한 Transformer 구조를 사용하지만, 더 적은 계산 자원과 메모리로 동작할 수 있도록 설계. 이를 위해 BERT의 attention 모듈을 경량화하고, 네트워크의 깊이를 조절하는 방법을 사용함.

성능 유지: MobileBERT는 모델 크기를 줄이면서도 성능 저하를 최소화하려는 목표를 가지고 있다. 이를 위해 Knowledge Distillation(지식 증류) 기법을 사용하여, 큰 BERT 모델로부터 작은 MobileBERT 모델을 학습시켰다. 이를 통해 MobileBERT는 BERT의 성능을 가깝게 유지할 수 있다.

빠른 추론 속도: 모바일 및 엣지 디바이스와 같은 제한된 환경에서 효율적으로 사용할 수 있도록 추론 속도가 최적화되었다. 이로 인해 실시간 자연어 처리 애플리케이션에서 유용하게 사용할 수 있다.

적용 분야: 모바일버트는 검색 엔진, 음성 비서, 챗봇, 언어 이해, 질의응답 시스템, 감정 분석 등 다양한 자연어 처리 작업에 활용될 수 있다.

- 보틀넷 구조:
  - 바일버트는 BERT의 모든 계층에서 입력 벡터를 보틀넥 계층을 거쳐 압축하는 과정을 추가함.
  - 보틀넥 계층: MobileBERT는 원래 BERT의 768차원의 벡터 대신, 4배 더 작은 128차원의 보틀넥 표현을 사용한다. 이를 통해 차원을 줄이고 연산량을 줄인다. 보틀넥 계층 이후에는 다시 768차원으로 확장되며, 이렇게 압축된 상태에서 효율적으로 정보를 처리하면서도 필요한 연산을 줄일 수 있다.

- 역보틀넷 (Inverted Bottlenet):
  - 작은 차원 -> 큰 차원 -> 작은 차원
  - 모바일버트는 Inverted Bottleneck(역 보틀넥) 구조를 사용한다. 이는 일반적인 보틀넥 구조와 달리, 입력 차원보다 중간 계층의 차원이 더 크게 확장된 후 다시 줄어드는 구조이다. 이 구조의 이점은 다음과 같다
    - 모델이 더 적은 차원에서 연산할 수 있어 계산 자원을 절약할 수 있다.
    - 차원이 줄어드는 중간 과정에서 중요한 특징들을 추출하면서도, 정보 손실을 최소화할 수 있다.
    - 이 구조는 MobileNetV2에서 도입되어 EfficientNet, GhostNet, ShuffleNetV2 등에 사용됨.

- BERT와의 차이점:
  - 기존 BERT 모델은 같은 차원의 벡터를 반복해서 처리하는 반면, MobileBERT는 보틀넥 구조를 사용하여 중간에 차원을 줄이면서 연산량을 줄이고, 다시 확장하여 BERT와 유사한 성능을 유지한다. 이는 모바일 및 엣지 환경에서의 실시간 자연어 처리에 매우 적합한 방식이다.

- Knowledge Distillation (지식 증류)
  - 모바일버트는 Knowledge Distillation(지식 증류) 기법을 사용하여 원래의 BERT 모델로부터 학습되었다. 즉, 큰 BERT 모델에서 학습된 지식을 MobileBERT로 전달하면서, 보틀넥 구조로 인한 성능 저하를 최소화하는 것이며 이를 통해 경량화된 아키텍처에서도 BERT의 성능을 최대한 유지할 수 있다.

### 요약
- 보틀넥 구조는 차원을 줄였다가 다시 확장하는 방식으로 연산량을 줄인다.
- 모바일버트는 역 보틀넥(Inverted Bottleneck) 구조를 사용하여 중요한 특징을 효율적으로 추출하고 계산 자원을 절약한다.
- 이 경량화된 구조는 모바일 및 엣지 디바이스에서도 높은 성능을 유지하면서도 적은 자원으로 작동할 수 있도록 설계되다.



## Knowledge Distillation

**지식 증류(Knowledge Distillation)**는 큰 모델(부모 모델, Teacher)에서 작은 모델(자식 모델, Student)로 지식을 전달하는 기법이다. 이 방법은 작은 모델이 부모 모델의 성능을 최대한 따라가면서도 더 적은 계산 자원과 메모리를 사용하도록 한다. 지식 증류의 핵심 개념은 부모 모델이 학습한 지식을 활용하여 자식 모델이 더 효과적으로 학습할 수 있게 한다는 점이다.

### 지식 증류의 학습 방법
- 부모 모델(Teacher Model) 학습:
  - 먼저, 큰 부모 모델(예: BERT)을 일반적인 방식으로 학습시킨다. 이 모델은 학습 데이터에 대해 매우 높은 성능을 가지며, 자식 모델이 학습하는 데 참고할 기준이 된다. 이 모델은 soft target을 생성하는 역할을 한다.

- 자식 모델(Student Model) 학습:
  - 자식 모델은 더 작은 모델이다. 이 모델이 부모 모델로부터 학습한 지식을 지식 증류를 통해 전달받는다. 자식 모델은 부모 모델이 학습한 내용을 최대한 재현하려고 한다.

### Loss 계산 방법
지식 증류에서 Loss를 계산하는 방법은 다음 두 가지 요소를 결합하여 진행된다

- Soft Target Loss (부모 모델의 예측과의 차이)
  - 이 Loss는 자식 모델의 출력과 부모 모델의 Soft Target 간의 차이를 계산한다. 이때, 일반적으로 **Kullback-Leibler divergence (KL divergence)**를 사용하여 두 확률 분포 간의 차이를 측정한다. 이 Loss는 부모 모델과 자식 모델이 같은 예측을 할 수 있도록 자식 모델을 학습시키는 역할을 한다.

  - KL Divergence Loss

    𝐿𝐾𝐷 = 𝑇2⋅𝐾𝐿(softmax(𝑧𝑠/𝑇),softmax(𝑧𝑡/𝑇))
  
  여기서,
  𝑧𝑠: 자식 모델의 출력 logits  
  𝑧𝑡: 부모 모델의 출력 logits  
  𝑇: 온도(Temperature), 부모 모델의 soft target을 더 부드럽게 만들어주는 파라미터   
  
  온도 𝑇는 Softmax 함수의 출력을 부드럽게 조정한다.
  𝑇가 크면 클래스 간 확률 분포가 더 고르게 나오고, 작으면 더 뚜렷하게 구분된다. 이를 통해 자식 모델은 다양한 클래스의 차이도 학습할 수 있다.

- Hard Target Loss (실제 정답 레이블과의 차이)
  - 두 번째로 자식 모델이 실제 정답 레이블(ground truth)과 얼마나 일치하는지를 확인하는 일반적인 Cross Entropy Loss를 계산한다. 이 Loss는 자식 모델이 실제로 정답 레이블에 맞는 예측을 할 수 있도록 한다.

  - Cross Entropy Loss:
    
    𝐿𝐶𝐸 = CrossEntropy(true label, student prediction)

- 최종 Loss 함수
  - 지식 증류에서 자식 모델의 최종 Loss는 Soft Target Loss와 Hard Target Loss의 가중합으로 계산된다. 이때 두 Loss의 비율을 적절히 조정하여 자식 모델이 학습을 균형 있게 할 수 있도록 한다.

  - 최종 Loss
  
    𝐿= 𝛼⋅𝐿𝐶𝐸 + 1−𝛼)⋅𝐿𝐾𝐷L

    여기서,
    α는 두 Loss의 가중치를 조절하는 하이퍼파라미터이다. 예를 들어, α를 0.5로 설정하면 자식 모델이 부모 모델의 Soft Target과 실제 레이블을 모두 균형 있게 학습할 수 있다.

### 참고
1. 하드 타겟 (Hard Target)  
하드 타겟은 일반적으로 우리가 알고 있는 **정답 레이블(ground truth label)**을 의미. 분류 문제에서 데이터의 각 샘플은 특정 클래스에 속하며, 해당 클래스는 확정된 값으로 표시됨. 하드 타겟은 각 샘플에 대해 오직 하나의 클래스가 정답으로 간주되며, 그 외의 클래스는 잘못된 것으로 처리된다.

- 예시: 이미지 분류 문제에서 고양이와 개를 분류하는 모델이라면,

| 클래스  | 확률 |
|---------|------|
| 고양이  | 1    |
| 개      | 0    |


하드 타겟은 항상 0 또는 1로만 표시되며, 정답 클래스는 1, 나머지 클래스는 0으로 표시된다.
모델이 이 값을 기반으로 학습하여 분류 작업을 수행한다.

2. 소프트 타겟 (Soft Target)  
소프트 타겟은 **부모 모델(Teacher Model)**이 각 클래스에 대해 예측한 확률 분포이다. 소프트 타겟은 모델이 해당 샘플이 어느 클래스에 속할 가능성이 있는지에 대한 추정 확률을 나타낸다.
소프트 타겟은 하드 타겟과 달리 여러 클래스에 대한 확률 분포를 포함하며, 모든 클래스에 대해 어느 정도의 가능성을 부여한다. 이 확률 분포는 예측의 불확실성을 반영하며, 정답 클래스뿐만 아니라 다른 클래스에 대한 정보도 자식 모델(Student Model)에 전달된다.

- 예시:고양이와 개를 분류하는 모델에서, 고양이 이미지를 처리했을 때 부모 모델이 다음과 같이 확률을 예측했다고 가정,

| 클래스  | 확률 |
|---------|------|
| 고양이  | 0.8    |
| 개      | 0.2    |

이 경우, 부모 모델은 해당 이미지가 고양이일 가능성이 높다고 판단하지만, 개일 가능성도 20% 정도 있다는 불확실성을 보여준다. 소프트 타겟은 하드 타겟처럼 특정 클래스에 대해 확정된 정답을 제공하지 않으며, 모델이 다양한 클래스에 대해 보다 유연하게 학습할 수 있도록 한다.

3. **온도(T)**와 소프트 타겟의 부드러움

지식 증류에서는 소프트 타겟의 부드러움을 조절하기 위해 **온도(Temperature, T)**라는 파라미터를 사용한다. Softmax 함수를 사용할 때, 이 온도를 조절하여 확률 분포를 더 부드럽게 만들 수 있다.
온도 𝑇를 높이면 클래스 간 확률 차이가 줄어들어 더 부드러운 확률 분포를 얻게 되고, 온도
𝑇가 낮으면 더 뚜렷한 예측이 이루어진다.

온도 𝑇=1

| 클래스  | 확률 |
|---------|------|
| 고양이  | 0.8    |
| 개      | 0.2    |

온도 𝑇=5

| 클래스  | 확률 |
|---------|------|
| 고양이  | 0.6    |
| 개      | 0.4    |

온도가 높아지면 모델이 더 많은 클래스에 대한 정보를 고려하게 된다. 자식 모델은 이런 부드러운 확률 분포를 사용해 부모 모델이 데이터에 대해 갖고 있는 불확실성을 학습할 수 있다.

4. 온도 값을 설정할 때의 고려 사항

- 높은 온도 값 (일반적으로 𝑇=2∼10):
  - 확률 분포를 부드럽게 만들어 자식 모델이 더 풍부한 정보를 얻을 수 있다.모든 클래스에 대한 정보를 조금씩 학습할 수 있게 되어 일반화 성능이 좋아질 수 있다. 하지만 너무 큰 온도 값은 클래스 간의 차이를 너무 줄여버려, 자식 모델이 중요한 특징을 놓칠 가능성도 있다.

- 낮은 온도 값 (𝑇 < 1):
  - 확률 분포가 더 뚜렷해지면서 부모 모델이 예측한 정답 클래스에만 집중하게 된다. 자식 모델이 특정 클래스에 과도하게 집중할 수 있으며, 결과적으로 일반화 성능이 떨어질 수 있다.

5. 온도 설정에 대한 실전 팁

- 실험적 설정: 지식 증류에서 적절한 온도 값은 실험적으로 설정하는 경우가 많다. 일반적으로 𝑇=2∼5 사이에서 시작하여 성능을 확인한 후, 필요에 따라 온도를 조정한다.
- 모델 및 데이터셋에 따라 다름: 온도 값은 모델의 크기나 학습할 데이터셋에 따라 달라질 수 있다. 예를 들어, 부모 모델이 더 복잡하고 성능이 뛰어날수록, 자식 모델이 이를 따라가야 하는 상황에서 더 큰 온도를 사용하는 것이 효과적일 수 있다.
- 자식 모델의 크기와도 관련: 자식 모델이 매우 작은 경우, 큰 온도 값을 사용할 때 부모 모델의 확률 분포를 지나치게 부드럽게 만드는 것보다는 적절한 중간 온도 값(예:𝑇=2)을 사용하는 것이 더 좋은 결과를 낳을 수 있다.

6. 요약

```python
# Mobile BERT 추론
from transformers import MobileBertTokenizer, MobileBertForMaskedLM
import torch

# 토크나이저 및 모델 불러오기
tokenizer = MobileBertTokenizer.from_pretrained('google/mobilebert-uncased')
model = MobileBertForMaskedLM.from_pretrained('google/mobilebert-uncased')

# 마스크한 문장 및 정답 문장을 각기 토크나이징
inputs = tokenizer("The capital of Korea is [MASK].", return_tensors="pt")
labels = tokenizer("The capital of Korea is Seoul.", return_tensors="pt")["input_ids"]

# 토크나이징 결과인 inputs(**kwargs 형식)와 lables를 모델에 입력
outputs = model(**inputs, labels=labels)
# 손실 추출
loss = outputs.loss
# 로짓 추출
logits = outputs.logits

# logits.argmax(-1)은 로짓 텐서의 마지막 축(axis)의 값들을 대상으로 argmax()를 적용
# logits.argmax(-1) 실행 결과의 첫번째 원소가 logits.argmax(-1)[0]

# 리스트 내포 [tokenizer.~.argmax(-1)[0]] 코드는
# logits.argmax(-1) 실행 결과의 첫번째 원소안의 구성 요소들을 하나씩 변수 i로 추출하여 item()으로 파이썬 숫자화하고
# 파이썬 숫자인 i.item()을 디코딩하고 빈 공백(스페이스)은 없앰

# 리스트 내포는 for 반복루프가 있기 때문에 처리 결과가 반복적으로 발생하는데
# 발생 결과에서 [1:-1]은 맨 처음 요소와 맨 마지막 요소를 제거
# 이렇게 처리된 결과를 join()으로 순차적으로 저장 후 print()로 출력
print(' '.join([tokenizer.decode(i.item()).replace(" ", "") for i in logits.argmax(-1)[0]][1:-1]))
```
