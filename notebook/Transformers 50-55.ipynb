# BART

## 목적

BART(Bidirectional and Auto-Regressive Transformers) 모델은 다양한 자연어 처리(NLP) 작업에서 높은 성능을 발휘하기 위해 설계된 **텍스트 생성 모델**. BART는 텍스트의 **복원 및 생성** 작업에 주로 사용되며, 그 목적은 다음과 같다:

- **텍스트 요약**: 긴 문서의 핵심 내용을 간결하게 요약.
- **기계 번역**: 한 언어의 텍스트를 다른 언어로 변환.
- **텍스트 복원**: 손상된 텍스트를 원래의 형태로 복원.
- **질문 생성**: 주어진 텍스트를 바탕으로 질문을 생성.

BART의 목표는 입력된 텍스트를 인코딩하여 중요한 정보를 추출하고, 이를 바탕으로 원하는 형태의 텍스트를 생성하는 것이다.

## 아키텍처

BART는 **Transformer 인코더-디코더 아키텍처**를 기반으로 한다. 이 아키텍처는 다음과 같은 특징을 갖는다:

- **인코더(Encoder)**: 입력 텍스트를 받아 고정된 크기의 잠재 표현(latent representation)으로 인코딩한다. BERT와 유사하게 양방향(Bidirectional)으로 텍스트를 처리하여 입력 시퀀스의 문맥을 학습.
  
- **디코더(Decoder)**: 인코더에서 생성된 잠재 표현을 받아 출력 시퀀스를 생성. GPT와 유사하게 순차적으로 텍스트를 생성하는 자동 회귀(Auto-Regressive) 방식을 사용.

![](https://dladustn95.github.io/assets/images/bart_figure1.png)

BART는 다음과 같은 방식으로 텍스트를 처리:

![](https://dladustn95.github.io/assets/images/bart_figure2.png)

1. **손상된 텍스트 입력**: 원래 텍스트를 마스킹, 순서 뒤바꿈, 삭제 등의 방법으로 손상시킨 후 입력으로 사용.
2. **인코딩**: 손상된 텍스트를 인코더에 입력하여 중요한 정보를 추출.
3. **디코딩**: 추출된 정보를 바탕으로 디코더에서 원래 텍스트를 복원하거나 새로운 텍스트를 생성.

## 실험

BART는 다양한 NLP 과제에서 실험적으로 검증됨. 주요 실험으로는 다음과 같다:

- **텍스트 요약**: CNN/DailyMail 데이터셋을 사용하여 뉴스 기사의 요약문을 생성하는 실험을 수행. BART는 긴 문서를 인코딩하여 중요한 정보를 추출한 후, 간결한 요약문을 생성.

- **기계 번역**: WMT 데이터셋을 사용하여 영어-독일어 번역 작업을 수행. BART는 양방향 인코딩과 자동 회귀 디코딩을 결합하여 높은 품질의 번역문을 생성.

- **텍스트 복원**: 손상된 텍스트를 원래 형태로 복원하는 실험을 통해 BART의 텍스트 복원 능력을 평가.

각 실험은 BART의 다양한 텍스트 생성 및 복원 작업에서의 성능을 평가하는 데 사용됨.

## 성능

BART는 여러 NLP 작업에서 최첨단 성능을 기록하며, 다음과 같은 장점을 보여줌:

- **텍스트 요약**: BART는 ROUGE 점수에서 다른 최신 모델들보다 높은 성능을 보여주며, 긴 문서를 효율적으로 요약할 수 있음을 입증.
- **기계 번역**: BART는 BLEU 점수에서 탁월한 성능을 보였으며, 다양한 언어 쌍에서 높은 품질의 번역문을 생성.
- **텍스트 복원**: 손상된 텍스트를 원래 형태로 정확하게 복원하는 능력이 뛰어남.

![](https://dladustn95.github.io/assets/images/bart_figure7.png)

## 결론

BART는 인코더-디코더 아키텍처를 기반으로 한 강력한 텍스트 생성 모델로, 텍스트 요약, 기계 번역, 텍스트 복원 등 다양한 NLP 작업에서 우수한 성능을 발휘. BART의 양방향 인코딩과 자동 회귀 디코딩 전략은 텍스트의 문맥을 정확하게 이해하고, 자연스러운 텍스트를 생성하는 데 중요한 역할을 함.
